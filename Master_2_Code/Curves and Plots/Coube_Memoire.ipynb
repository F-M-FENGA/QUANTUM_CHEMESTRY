```python
import matplotlib.pyplot as plt
import numpy as np

# Données
methodes = ["SSVQE", "TD-DFT", "qEOM"]
energies = [2.360, 3.208, 4.914]  # eV

# Positions pour les barres
x = np.arange(len(methodes))
bar_width = 0.6

# Création du graphique
plt.figure(figsize=(7,5))
bars = plt.bar(x, energies, bar_width, color=['#4CAF50', '#FF9800', '#2196F3'], alpha=0.8)

# Axes et labels
plt.ylabel("Énergie d'excitation ΔE (eV)", fontsize=12)
plt.xticks(x, methodes, fontsize=11)
plt.ylim(0, max(energies) + 1)

# Légende et titre
plt.legend(bars, methodes, title="Méthodes", loc="upper right")
plt.title("Comparaison des énergies d'excitation", fontsize=14, fontweight='bold')

# Ajustement et affichage
plt.tight_layout()
plt.savefig("energies_1.png", dpi=300)  # <-- sauvegarde en PNG
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# --- Données ---
methodes = ["SSVQE", "TD-DFT", "QEOM"]
energies = np.array([2.360, 3.208, 4.914])  # eV
longueurs_onde = np.array([525.43, 386.40, 252.34])  # nm

# Valeurs expérimentales
E_exp = 2.460  # eV
lambda_exp = 504  # nm

# Écarts absolus
ecarts_energie = np.abs(energies - E_exp)
ecarts_lambda = np.abs(longueurs_onde - lambda_exp)

# --- Graphique 1 : Énergies ---
plt.figure(figsize=(6,4))
bars = plt.bar(methodes, energies, color=['#4CAF50', '#FF9800', '#2196F3'], alpha=0.8)
plt.axhline(E_exp, color='red', linestyle='--', linewidth=1.5, label=f'Expérimental ({E_exp} eV)')
plt.ylabel("Énergie d'excitation ΔE (eV)")
plt.title("Énergies d'excitation pour BODIPY")
plt.legend()
plt.tight_layout()
plt.savefig("energies.png", dpi=300)
plt.close()

# --- Graphique 2 : Longueurs d'onde ---
plt.figure(figsize=(6,4))
bars = plt.bar(methodes, longueurs_onde, color=['#4CAF50', '#FF9800', '#2196F3'], alpha=0.8)
plt.axhline(lambda_exp, color='red', linestyle='--', linewidth=1.5, label=f'Expérimental ({lambda_exp} nm)')
plt.ylabel("Longueur d'onde λ (nm)")
plt.title("Longueurs d'onde pour BODIPY")
plt.legend()
plt.tight_layout()
plt.savefig("longueurs_onde.png", dpi=300)
plt.close()

# --- Graphique 3 : Écarts absolus ---
plt.figure(figsize=(6,4))
bars = plt.bar(methodes, ecarts_energie, color=['#4CAF50', '#FF9800', '#2196F3'], alpha=0.8)
plt.ylabel("Écart absolu ΔE (eV)")
plt.title("Écarts absolus des énergies d'excitation par rapport à l'expérimental")
plt.tight_layout()
plt.savefig("ecarts_absolus.png", dpi=300)
plt.close()

print("Les trois graphiques ont été générés et sauvegardés : 'energies.png', 'longueurs_onde.png', 'ecarts_absolus.png'.")
```

```python
import numpy as np

# --- Données simulées pour plusieurs transitions ---
# Exemple : SSVQE, TD-DFT, QEOM pour 3 transitions différentes (ΔE en eV)
methodes = ["SSVQE", "TD-DFT", "QEOM"]
energies_calc = np.array([
    [2.360],  # SSVQE
    [3.208],  # TD-DFT
    [4.914]   # QEOM
])

# Valeurs expérimentales correspondantes
energies_exp = np.array([2.460, 2.470, 2.490])

# --- Calcul des écarts absolus ---
ecarts_abs = np.abs(energies_calc - energies_exp)

# --- Calcul MAE et RMSE par méthode ---
MAE = np.mean(ecarts_abs, axis=1)
RMSE = np.sqrt(np.mean(ecarts_abs**2, axis=1))

# Affichage
for i, methode in enumerate(methodes):
    print(f"{methode} : MAE = {MAE[i]:.3f} eV, RMSE = {RMSE[i]:.3f} eV")
```
